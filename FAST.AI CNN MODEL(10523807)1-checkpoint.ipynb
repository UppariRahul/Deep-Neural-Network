```python
# Fastai Libraries and these libraries are being used to be readily avaialbe and fast productive

import fastai
from fastai import *
from fastai.vision import *  # this library is used to develop image classification models
```

```python
# storing the data from the source into a variable

path = pathlib.Path('E:\Thesis\dataset')
path
```

```python
# bs to define the batch size to define number of training samples used in one iteration

bs = 8

# Data is stored in 29 different folders name and To extract and classify the images data into right categories, Fastai library 
# gives unique function name ImageDataBunch where it automatically assigns the labels to the data as they are stored in their 
# respective folders.
# Normalizing the data using Pretrained Imagenet model.

data = ImageDataBunch.from_folder(path=path,bs=bs,ds_tfms=get_transforms(),size=200,valid_pct=0.2)
data.normalize(imagenet_stats)
```

```python
# To Print all the classes respective to the data stored in the folders.

print(data.classes)
data.c
```

```python
# To show the Image data present in the folders.

data.show_batch(rows=3)
```

```python
# defining a function for Conv layer with Kernel, Stride and Padding Value which will same for the entire network.

def conv(ni,nf): return nn.Conv2d(ni, nf, kernel_size=3, stride=2, padding=1)
```

```python
model = nn.Sequential(      # to create a constructor in which layers are processed as they are specified in it.
    conv(3, 8), # 14
    nn.BatchNorm2d(8),      #to normalize the input data and reduce the internal coveriate shift
    nn.ReLU(),              ## this layer is used to convert negative values with zero and use only positive values
    conv(8, 16), # 7
    nn.BatchNorm2d(16),
    nn.ReLU(),
    conv(16, 32), # 4
    nn.BatchNorm2d(32),
    nn.ReLU(),
    conv(32, 16), # 2
    nn.BatchNorm2d(16),
    nn.ReLU(),
    conv(16, 10), # 1
    nn.BatchNorm2d(10),
    Flatten()     # remove (1,1) grid # this converts our 3D feature maps to 1D feature vectors
)
```

```python
# creating a learner model which will be later used to train the model and do predictions

learn = Learner(data, model, loss_func = nn.CrossEntropyLoss(), metrics=accuracy)
```

```python
# This functions shows the entire stack of neural network layer which we have defined for our model training.
learn.summary()
```

```python
# this function helps to find the best learning rate which can be used to train the model
# on that learning rate to get best results.

learn.lr_find(end_lr=100)
```

```python
# It helps to plot the learning rate to visualize it and select the best learning rate from the graph 
# in the below graph the best learning rate is 1e+01.

learn.recorder.plot()
```

```python
# Here we are training/ fitting the model and specifying the epoch size = 3 and learning rate = 1e+01 select from the above plot
# fit_one_cycle is used to train the model with cylical learning rates based on the iteration defined.
# And also it helps to interpret the train_loss, valid_loss, accuracy, time.

learn.fit_one_cycle(3, max_lr = 1e+01)
```

```python
# These lines of code helps to understand the losses, helps to create confusion matrix and which is used to visualize
# incorrect images predicted.

interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
# This line of code prints the prediction/ loss/ probability values


interp.plot_top_losses(9,figsize=(15,11))
```

```python
# To plot the confusion matrix and interpret the losses and incorrect category predicted.
# To identify the top losses between different categories

interp.plot_confusion_matrix(figsize=(12,12), dpi=60)
```

```python
# This line of code gives the text interpretion of confusion matrix and helps to understand between which values the model is 
# more confused to predict the actual value.

interp.most_confused(min_val=2)
```

```python
# Residula Learning Block (ResBlock())
```

```python
#It contains two conv layers (a conv_layer is Conv2d, ReLU, batch norm),
#so create two of those and then in forward we go conv1(x) and conv2 of that and then add x.
#ResBlock() increases the size of the network layer when compared with actual network layer.

class ResBlock(nn.Module):
    def __init__(self, nf):
        super().__init__()
        self.conv1 = conv_layer(nf,nf)
        self.conv2 = conv_layer(nf,nf)
        
    def forward(self, x): return x + self.conv2(self.conv1(x))
```

```python
model = nn.Sequential(
    conv(3, 8),
    res_block(8),  #it takes the input of the above conv layer and adds the value to next layer of the conv.
    conv(8, 16),
    res_block(16),
    conv(16, 32),
    res_block(32),
    conv(32, 16),
    res_block(16),
    conv(16, 10),
    Flatten()
)
```

```python
# Refactoring the entire neural network and build the whole network only with convolution layer and flatten layer to 
# check if it helps to increase the model accuracy

def conv_and_res(ni,nf): return nn.Sequential(conv2d(ni, nf), res_block(nf))
```

```python
model = nn.Sequential(
    conv_and_res(3, 8),
    conv_and_res(8, 16),
    conv_and_res(16, 32),
    conv_and_res(32, 16),
    conv(16, 10),
    Flatten()
)
```

```python
learn = Learner(data, model, loss_func = nn.CrossEntropyLoss(), metrics=accuracy)
```

```python
learn.lr_find(end_lr=100)
learn.recorder.plot()
```
