```python
import tensorflow as tf            # to use Tensorflow API for smooth use of CPU and GPU
import os                       #import dataset from the local system
import cv2                      # Used to develop real time computer vision applications
from tqdm import tqdm           # to run nested loop
import matplotlib.pyplot as plt # to plot graphs
from tensorflow.keras.models import Sequential                #to define the order of the layers specified in the constructor
from tensorflow.keras.layers import Conv2D, MaxPooling2D                 # To create layers which is used to extract features
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten      # to develop which can understand complex images
# more info on callbakcs: https://keras.io/callbacks/ model saver is cool too.
from tensorflow.keras.callbacks import TensorBoard               # To visualize the model in Tensorboard Framework
import pickle                   # for Serializing and De-serializing python object structures
import time 
```

```python
# creating the data for model training and predictions
training_data = []

def create_training_data():
    for category in categories:  # for all 29 folders.

        path = os.path.join(DATADIR,category)  # create path 
        class_num = categories.index(category)  # get the classification

        for img in tqdm(os.listdir(path)):  # iterate over each image 
            try:
                img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE)  # convert to array
                new_array = cv2.resize(img_array, (Img_size, Img_size))  # resize to normalize data size
                training_data.append([new_array, class_num])  # add this to our training_data
            except Exception as e:  # in the interest in keeping the output clean...
                pass
            #except OSError as e:
            #    print("OSErrroBad img most likely", e, os.path.join(path,img))
            #except Exception as e:
            #    print("general exception", e, os.path.join(path,img))

create_training_data()

print(len(training_data))
```

```python
import random

# to randomly shuffle the training data so it is not biased or train on the same type of image.
random.shuffle(training_data)
```

```python
for sample in training_data[:10]:
    print(sample[1])
```

```python
import numpy as np

# to store the feature data in X variable and Labels of the dataset as per folder name in y variable
X = []
y = []

for features,label in training_data:
    X.append(features)
    y.append(label)

print(X[0].reshape(-1, Img_size, Img_size, 1))

# used to give an new shape to array without changing its data. The new shope should be compatibele to the original data.
# reshape is used to convert 2D array into single vector column.

X = np.array(X).reshape(-1, Img_size, Img_size, 1)
```

```python
import pickle

# to serialize and de-serialize the object structure and store into the file/database which can used later to avoid the whole
# above mentioned process.
pickle_out = open("X.pickle_keras_ASL","wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle_keras_ASL","wb")
pickle.dump(y, pickle_out)
pickle_out.close()
#We can always load it in to our current script, or a totally new one by doing:

# pickle_in = open("X.pickle","rb")
# X = pickle.load(pickle_in)

# pickle_in = open("y.pickle","rb")
# y = pickle.load(pickle_in)
```

```python
#We can always load it in to our current script, or a totally new one by doing:

pickle_in = open("X.pickle_keras","rb")
X = pickle.load(pickle_in)

pickle_in = open("y.pickle_keras","rb")
y = pickle.load(pickle_in)
y = np.array(y)

X = X/255.0
```

```python
# This varialbe is defined as a name to the log files.
NAME = 'KERAS_ASL_CNN'
```

```python
model = Sequential() # to create a constructor in which layers are processed as they are specified in it.

model.add(Conv2D(32, (3, 3), input_shape=X.shape[1:])) # This layer is used to extract low-level features using kernel, stride and padding
model.add(Activation('relu')) # this layer is used to convert negative values with zero and use only positive values
model.add(MaxPooling2D(pool_size=(2, 2))) # to conver and extract maximum feature values and high-level features from the image

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
model.add(Dense(128))
model.add(Activation('relu'))

model.add(Dense(29)) # this layer is used to feed all the outputs from the previous layer to all its neuron.
model.add(Activation('softmax')) # used in the output layer trying to attain the probabilities to define the class of each input

tensorboard = TensorBoard(log_dir="logs/{}".format(NAME)) #to store the logs which is can be used for visualization.

model.compile(loss='sparse_categorical_crossentropy', #defined as categorical crossentropy with integer targets
              optimizer='adam',  #optimizer is used to change the weights and learning rate
              metrics=['accuracy'], #to display the model results.
              )
```

```python
# model is trained with the specified number of epoch which give the model accuracy, train losses, validation losses
# and callback function is used to visualize the model with plots.

history = model.fit(X, y,
          batch_size=8,
          epochs=3,
          validation_split=0.3,
          callbacks=[tensorboard])
```

```python
# This line of code is used to save the model in the file/ database which can be used again 
# instead executing the whole model and training it again and again.

model.save('KERAS_ASLCNN.model')
```

```python
# This line of code is used to load the model which we have saved in the file database

model = tf.keras.models.load_model("KERAS_ASLCNN.model")
```

```python
# This line of code is used to plot the train loss of the model

import matplotlib.pyplot as plt
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(loss) + 1)
plt.plot(epochs, loss, color='red', label='Training loss')
plt.plot(epochs, val_loss, color='green', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# This line of code is used to plot accuracy values of the model.

accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
plt.plot(epochs, accuracy, color='red', label='Training acc')
plt.plot(epochs, val_accuracy, color='green', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# We are specifying the required layers to build the custom Neural Network Layers.

dense_layers = [0, 1, 2]
layer_sizes = [32, 64, 128]
conv_layers = [1, 2, 3]
```

```python
NAME = 'Tune_ASL_CNN'
```

```python
# In this set of code using the aboue custom values we are executing a for loop so that our Network Layer is executed for every
# value till it completes the cycle of execution which is return can be termed as Parameter Tunning.

for dense_layer in dense_layers:
    for layer_size in layer_sizes:
        for conv_layer in conv_layers:
            NAME = "{}-conv-{}-nodes-{}-dense-{}".format(conv_layer, layer_size, dense_layer, int(time.time()))
            print(NAME)

            model1 = Sequential()

            model1.add(Conv2D(layer_size, (3, 3), input_shape=X.shape[1:]))
            model1.add(Activation('relu'))
            model1.add(MaxPooling2D(pool_size=(2, 2)))

            for l in range(conv_layer-1):
                model1.add(Conv2D(layer_size, (3, 3)))
                model1.add(Activation('relu'))
                model1.add(MaxPooling2D(pool_size=(2, 2)))

            model1.add(Flatten())
            for _ in range(dense_layer):
                model1.add(Dense(layer_size))
                model1.add(Activation('relu'))

            model1.add(Dense(29))
            model1.add(Activation('softmax'))
            
# Below output is the stack of layers which is been created with the help of for loop
```

```python
tensorboard = TensorBoard(log_dir="logs/{}".format(NAME)) #to store the logs which is can be used for visualization.
```

```python
# In this line of code the model is complied 
model.compile(optimizer='adam',  # optimizer is used to change the weights and learning rate to reduce losses
              loss='sparse_categorical_crossentropy', #defined as categorical crossentropy with integer targets
              metrics=['accuracy'])  # to display the model results
```

```python
# model is trained with the specified number of epoch which give the model accuracy, train losses, validation losses
# and callback function is used to visualize the model with plots.

model.fit(X, y, epochs=8, batch_size = 8 , validation_split=0.3, callbacks = [tensorboard])
```
